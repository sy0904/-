## ブロックマッチングの算法を自分で書いて確かめましょう．
1. [Depth from Stereo](https://github.com/IntelRealSense/librealsense/blob/master/doc/depth-from-stereo.md)をよく読んでください．

Q06: OpenCV の関数を使って推定した視差とどこが異なるか．どのような工夫を施すと，StereoBM または StereoSGBM に近い推定結果や計算時間を達成できるか．

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# パラメータ
disparities = 64
block = 15
half = block // 2

# グレースケール画像に変換
cvleft = cv2.cvtColor(left, cv2.COLOR_RGB2GRAY).astype(np.float32)
cvright = cv2.cvtColor(right, cv2.COLOR_RGB2GRAY).astype(np.float32)

# 視差画像
disparity = np.zeros(cvleft.shape, dtype=np.float32)

# SSDブロックマッチング
for i in range(half, cvleft.shape[0] - half):
    for j in range(half + disparities, cvleft.shape[1] - half):

        best_ssd = float("inf")
        best_disp = 0

        # 左右画像を比較
        for d in range(disparities):

            left_block = cvleft[
                i-half:i+half+1,
                j-half:j+half+1
            ]

            right_block = cvright[
                i-half:i+half+1,
                j-d-half:j-d+half+1
            ]

            ssd = np.sum((left_block - right_block) ** 2)

            if ssd < best_ssd:
                best_ssd = ssd
                best_disp = d

        disparity[i, j] = best_disp

# 表示
plt.figure(figsize=(10,6))
plt.imshow(disparity, cmap="gray", vmin=0, vmax=disparities)
plt.colorbar(label="Disparity")
plt.title("SSD Block Matching")
plt.show()
```

```python
import numpy

fx = 942.8        # lense focal length
baseline = 54.8   # distance in mm between the two cameras
disparities = 64  # num of disparities to consider
block = 15        # block size to match
units = 0.001     # depth units

for i in range(block, left.shape[0] - block - 1):
    for j in range(block + disparities, left.shape[1] - block - 1):
        ssd = numpy.empty([disparities, 1])

        # calc SSD at all possible disparities
        l = left[(i - block):(i + block), (j - block):(j + block)]
        for d in range(0, disparities):
            r = right[(i - block):(i + block), (j - d - block):(j - d + block)]
            ssd[d] = numpy.sum((l[:,:]-r[:,:])**2)

        # select the best match
        disparity[i, j] = numpy.argmin(ssd)

# Convert disparity to depth
depth = np.zeros(left.shape[:2]).astype(float)
depth[disparity > 0] = (fx * baseline) / (units * disparity[disparity > 0])
```
